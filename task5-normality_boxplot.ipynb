```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

project_root = Path("/Users/miki_przygoda/Projects/GitHub-Projects/Maths Data Script")
csv_path = project_root / "full_dataset" / "full_dataset_20260313_134447_4ccad990.csv"

if not csv_path.exists():
    raise FileNotFoundError(f"CSV not found: {csv_path}")

print(f"Reading CSV: {csv_path}")
df = pd.read_csv(csv_path)
df_filtered = df[df['Member_Name'].astype(str).str.strip().str.lower() != 'mikolaj'].copy()

# Prepare variables for Model 3
cols = ['Process_Count', 'CPU_Usage_Pct', 'Execution_Time_ms']
for col in cols:
    df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce')
df_filtered = df_filtered.dropna(subset=cols)

X = df_filtered[['Process_Count', 'CPU_Usage_Pct']]
y = df_filtered['Execution_Time_ms']

# Fit model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Residuals (Normality Check)', fontsize=14)
plt.xlabel('Residual Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.savefig('model3_residuals_histogram.png', dpi=300)
plt.close()

print(f"Residual mean: {np.mean(residuals):.10f}")
print(f"Residual std dev: {np.std(residuals):.10f}")
```
