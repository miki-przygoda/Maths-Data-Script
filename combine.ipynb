```python
import uuid
from pathlib import Path
from datetime import datetime
import pandas as pd

data_root = Path("data")
output_dir = Path("full_dataset")
output_dir.mkdir(parents=True, exist_ok=True)

csv_files = sorted(data_root.rglob("*.csv"))

if not csv_files:
    raise FileNotFoundError(f"No CSV files found under: {data_root.resolve()}")

frames = []
for f in csv_files:
    try:
        df = pd.read_csv(f)

        rel_parts = f.relative_to(data_root).parts
        folder_name = rel_parts[0] if len(rel_parts) > 1 else "root"

        df["Member_Name"] = folder_name
        frames.append(df)
    except Exception as e:
        print(f"Skipping {f}: {e}")

if not frames:
    raise RuntimeError("CSV files were found, but none could be read.")

combined = pd.concat(frames, ignore_index=True, sort=False)

uid = f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_{uuid.uuid4().hex[:8]}"
output_file = output_dir / f"full_dataset_{uid}.csv"
combined.to_csv(output_file, index=False)

print(f"Combined {len(frames)} files into:")
print(output_file.resolve())
print(f"Rows: {len(combined):,} | Columns: {len(combined.columns)}")
```
